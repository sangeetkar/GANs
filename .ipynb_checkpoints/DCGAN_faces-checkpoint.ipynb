# Generating new faces with **Generative Adversial Nets (GANs)**


## Introduction

I find **Generative Adversarial Nets framework (GANs)**, [proposed](https://arxiv.org/pdf/1406.2661.pdf) by Ian Goodfellow and his colleagues at University of Montreal, very fascinating. This novel technique relies on **pitching two different deep neural nets against each other** to achieve state-of-the-art results in estimating generative models:

 - **Generator (G)** which takes a random input and tries to generate an output (fake data) close to the real data (and consequently fool the other DNN)
 - **Discriminator (D)** which takes as input real data and generated data (fake data) and tries to distinguish one from the other

*Both these DNNs try to outdo each other and thus keep improving themselves in their function.*

Since a picture is a lot better than a thousand words, here's a diagram that explains it really well:

![GAN diagram](./images/GAN.png)

## Current Task

Here I'll apply this technique to the [Labeled Faces in the Wild (LFW)](http://vis-www.cs.umass.edu/lfw/) dataset and will use the generator to generate some novel faces. How exciting!

## Approach

### Architecture

Since we're dealing with images here, Convolutional Neural Nets (CNNS) are expected to be more effective. Unsurprisingly, Alec Radford et al., in 2016, [proposed](https://arxiv.org/pdf/1511.06434.pdf) **Deep Convolutional Generative Adversarial Nets (DCGANs)** which use CNNs (& Deconvolutional layers) as the architecture for Generator & Discriminator nets.  


### Deep Learning Library

I'll use the [Pytorch Library](http://pytorch.org/) in this project. Although I like [Tensorflow](https://www.tensorflow.org/), I'm increasingly getting more inclined towards pytorch because it's quite similar to python libraries like [Numpy](http://www.numpy.org/) and debugging is a lot easier than Tensorflow.



## Implementation

We start with importing the necessary libraries/ packages.

```python
# Imports
import torch
import torch.nn as nn
import torch.optim as optim
from torch.autograd import Variable
import torchvision
from torchvision import datasets, transforms

#Check if GPU is available. We'll use this flag to use GPU for training if available
has_gpu = torch.cuda.is_available()
```

Here I define the transforms I wish to apply on the images:

1. Scale to 64X64 (to reduce training time)
2. Center Crop 
3. Convert from image format to tensors
4. Normalize 

Later we define the dataset (a folder containing the images) and the loader (a python generator that we are going to use during training.

Let's take a look at some sample images from the dataset:

![Sample images](./images/real_samples.png)

The black uneven borders were present in the dataset (as a result of aligning the faces using deep-funneling)

```python
#preprocess images. We'll resize images to 64 X 64 size to reduce training time

image_size = 64
data_folder = './data'

# Transforms to be applied to the input images : Resize, Center-Crop to 64X64, convert to tensor and normalize 
transform = transforms.Compose([
    transforms.Scale(image_size),
    transforms.CenterCrop(image_size),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

# Define Dataset (here a folder containing the images) and specifying the transforms to be applied
faces = datasets.ImageFolder(data_folder, transform=transform)

batch_size=64
num_workers=4

# Prepare the loader (a python generator)
dataloader = torch.utils.data.DataLoader(faces, batch_size=batch_size, shuffle=True, num_workers=num_workers)
```

Here's an utility function that we'll use to initialize the weights of Convolutional & Batch Normalization layers

```python
# Initialize the Convolutional & Batch Normalization layers with weights from a normal distribution

def weights_init(m):
    classname = m.__class__.__name__
    if classname.find('Conv') != -1:
        m.weight.data.normal_(0.0, 0.02)
    elif classname.find('BatchNorm') != -1:
        m.weight.data.normal_(1.0, 0.02)
        m.bias.data.fill_(0)
```

According to the DCGAN paper, here's how a DCGAN generator looks like: 
 
![DCGAN Generator](https://blog.openai.com/content/images/2017/02/gen_models_diag_1.svg)

```python
# Here we define the Generator

class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        nz = 100
        self.main = nn.Sequential(
            nn.ConvTranspose2d(nz, 64 * 8, 4, 1, 0, bias=False),
            nn.BatchNorm2d(64 * 8),
            nn.ReLU(True),
            nn.ConvTranspose2d(64 * 8, 64 * 4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(64 * 4),
            nn.ReLU(True),
            nn.ConvTranspose2d(64 * 4, 64 * 2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(64 * 2),
            nn.ReLU(True),
            nn.ConvTranspose2d(64 * 2, 64, 4, 2, 1, bias=False),
            nn.BatchNorm2d(64),
            nn.ReLU(True),
            nn.ConvTranspose2d(64, 3, 4, 2, 1, bias=False),
            nn.Tanh()
        )
        
    def forward(self, inp):
        output = self.main(inp)
        return output
    
netG = Generator()
netG.apply(weights_init)
```

The Discriminator is like a normal Convolutional Neural Net. 
During my first attempt, the Discriminator gradually turned too powerful and started to overfit. That's why I have added some Drop-out layers. The fight has to be fair, right?

```python
# And here we define the Discriminator

class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.main = nn.Sequential(
            nn.Conv2d(3, 64, 4, 2, 1, bias=False),
            nn.LeakyReLU(0.2, True),
            nn.Conv2d(64, 64 * 2, 4, 2, 1, bias=False),
            nn.Dropout2d(0.4, True),
            nn.BatchNorm2d(64 * 2),
            nn.LeakyReLU(0.2, True),
            nn.Conv2d(64 * 2, 64 * 4, 4, 2, 1, bias=False),
            nn.Dropout2d(0.4, True),
            nn.BatchNorm2d(64 * 4),
            nn.LeakyReLU(0.2, True),
            nn.Conv2d(64 * 4, 64 * 8, 4, 2, 1, bias=False),
            nn.Dropout2d(0.4, True),
            nn.BatchNorm2d(64 * 8),
            nn.LeakyReLU(0.2, True),
            nn.Conv2d(64 * 8, 1, 4, 1, 0, bias=False),
            nn.Sigmoid()
        )
        
    def forward(self, inp):
        output = self.main(inp)
        return output.view(-1, 1)
    
netD = Discriminator()
netD.apply(weights_init)
```

```python
# Loss function
criterion = nn.BCELoss()

# Variables to be used in training
input = torch.FloatTensor(batch_size, 3, image_size, image_size)
noise = torch.FloatTensor(batch_size, 100, 1, 1)
fixed_noise = torch.FloatTensor(batch_size, 100, 1, 1).normal_(0, 1)
label = torch.FloatTensor(batch_size)

real_label = 1
fake_label = 0

if has_gpu:
    netD.cuda()
    netG.cuda()
    criterion.cuda()
    input, label = input.cuda(), label.cuda()
    noise, fixed_noise = noise.cuda(), fixed_noise.cuda()

input = Variable(input)
noise = Variable(noise)
fixed_noise = Variable(fixed_noise)
label = Variable(label)

optimizerG = optim.Adam(netG.parameters(), lr=0.0002, betas=(0.5, 0.999))
optimizerD = optim.Adam(netD.parameters(), lr=0.0002, betas=(0.5, 0.999))
```

```python
D_loss_vals = []
G_loss_vals = []

n_epochs = 100
for epoch in range(n_epochs):
    for i, (images, _) in enumerate(dataloader, 0):
        
        #train discriminator
        netD.zero_grad()
        batch_size1 = images.size(0)
        input.data.resize_(images.size()).copy_(images)
        label.data.resize_(batch_size1).fill_(real_label)

        output = netD(input)
        real_loss = criterion(output, label)
        real_loss.backward()
        D_x = output.data.mean()

        
        noise.data.resize_(batch_size1, 100, 1, 1).normal_(0,1)
        fake = netG(noise)
        label.data.fill_(fake_label)
        output = netD(fake.detach())
        fake_loss = criterion(output, label)
        fake_loss.backward()
        D_G_z1 = output.data.mean()
        
        D_loss = real_loss + fake_loss
        D_loss_vals.append(D_loss.data[0])
        optimizerD.step()

        #train generator
        netG.zero_grad()
        label.data.fill_(real_label)
        output = netD(fake)
        G_loss = criterion(output, label)
        G_loss.backward()
        D_G_z2 = output.data.mean()
        optimizerG.step()
        G_loss_vals.append(G_loss.data[0])
        
        if i % 100 == 0:
            print('[%d/%d][%d/%d] Loss_D: %.4f Loss_G: %.4f D(x): %.4f D(G(z)): %.4f / %.4f'
              % (epoch, n_epochs, i, len(dataloader),
                 D_loss.data[0], G_loss.data[0], D_x, D_G_z1, D_G_z2))
            torchvision.utils.save_image(images,
                    '%s/real_samples.png' % './generated',
                    normalize=True)
    netG.eval()
    fake = netG(fixed_noise)
    netG.train()
    torchvision.utils.save_image(fake.data,
           '%s/fake_samples_epoch_%03d.png' % ('./generated', epoch),
            normalize=True)
        
```

```python
#Save the Discriminator & Generator model weights for future training or inference

torch.save(netG.state_dict(), '%s/netG_epoch_%d.pth' % ('models', epoch))
torch.save(netD.state_dict(), '%s/netD_epoch_%d.pth' % ('models', epoch))
```

```python
obs = len(D_loss_vals)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

plt.figure(figsize=(12,8))
plt.plot(range(obs), D_loss_vals, label='D Loss')
plt.plot(range(obs), G_loss_vals, label='G Loss')
plt.legend(loc='best')
plt.xlabel('Batch no.')
plt.ylabel('Loss')
plt.savefig('./images/loss.png')
plt.show()
```
